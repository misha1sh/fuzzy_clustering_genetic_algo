```python
def labels_to_clusters(labels):
    cnt = max(labels)
    clusters = [[] for _ in range(cnt)]
    for j, jvalue in enumerate(labels):
        clusters[jvalue].append(j)

def clusters_to_labels(clusters):
    cnt = sum(len(i) for i in clusters)
    labels = np.zeros((cnt))
    for i, cluster in enumerate(clusters):
        for j in cluster:
            labels[j] = i
    return labels.astype(int)
    
```

```python
from scipy.spatial.distance import euclidean

def DaviesBouldin(X, y, labels):
    n_cluster = len(np.bincount(labels))
    cluster_k = [X[labels == k] for k in range(n_cluster)]
    centroids = [np.mean(k, axis = 0) for k in cluster_k]

    # calculate cluster dispersion
    S = [np.mean([euclidean(p, centroids[i]) for p in k]) for i, k in enumerate(cluster_k)]
    Ri = []

    for i in range(n_cluster):
        Rij = []
        # establish similarity between each cluster and all other clusters
        for j in range(n_cluster):
            if j != i:
                r = (S[i] + S[j]) / euclidean(centroids[i], centroids[j])
                Rij.append(r)
         # select Ri value of most similar cluster
        Ri.append(max(Rij)) 

    # get mean of all Ri values    
    dbi = np.mean(Ri)

    return dbi

    
DaviesBouldin(X, y, clusters_to_labels(clusters))
```

```python
def triangle_form(domain, points_count, left, middle, right):
    x = np.linspace(domain[0], domain[1], num=points_count)
    y = np.zeros_like(x)
    between_left_and_middle = np.bitwise_and(left <= x, x <= middle)

    if left >= middle:
        y[between_left_and_middle] = 1
    else:
        y[between_left_and_middle] = (x[between_left_and_middle] - left) / (middle - left)

    between_middle_and_right = np.bitwise_and(middle <= x, x <= right)

    if middle >= right:
        y[between_middle_and_right] = 1
    else:
        y[between_middle_and_right] = (right - x[between_middle_and_right]) / (right - middle)

    return y


def or_rule(a, b):
    return np.maximum(a, b)

def and_rule(a, b):
    return np.minimum(a, b)

def not_rule(a):
    return 1 - a

a = triangle_form([0, 1], 100, 0.1, 0.2, 0.5)
b = triangle_form([0, 1], 100, 0.3, 0.5, 0.8)

plt.plot(a, label='a')
plt.plot(b, label='b')
plt.legend()
plt.show()

plt.plot(or_rule(a, b), label='a or b')
plt.legend()
plt.show()

plt.plot(and_rule(a, b), label='a and b')
plt.legend()
plt.show()

plt.plot(not_rule(a), label='not a')
plt.legend()
plt.show()

# np.maximum(
#     np.array([[1, 2, 3],
#              [4, 5, 6]]),
#     np.array([[0, 4, 2],
#               [7, 2, 1]])
# )
```

```python
from collections import defaultdict

def fuzzify(fuzzy_system, variables):
    fuzzy_variables = {}
    for fuzzy_form_name, fuzzy_form in fuzzy_system['fuzzy_forms'].items():
        if fuzzy_form_name not in variables:
            continue

        domain = fuzzy_form['domain']
        for value_name, form_value in fuzzy_form['values'].items():
            x = ((variables[fuzzy_form_name] - domain[0]) / (domain[1] - domain[0] + 1) * len(form_value)).astype(int)
            fuzzy_variables[fuzzy_form_name + "_" + value_name] = form_value[x]
    return fuzzy_variables


def centroid_aggregation(fuzzy_system, values):
    aggregated_values = {}
    for value_name, value_form in values.items():
        x = value_form
        # domain = fuzzy_system['fuzzy_forms'][value_name]['domain']
        domain = (0, 1)
        y = np.linspace(domain[0], domain[1], x.shape[-1])

        ssum = np.sum(x, axis=-1)
        left = np.sum((x * y), axis=-1)
        # aggregated_values[value_name] = np.divide(left, ssum)
        aggregated_values[value_name] = np.divide(left, ssum,
                 out=np.zeros_like(left), where=ssum!=0)

    return aggregated_values

def run_fuzzy_rules(fuzzy_system, fuzzy_variables):
    res = dict()
    for rule in fuzzy_system['rules']:
        rule_strength, out_variable, out_variable_value = rule(fuzzy_variables)
        # out_variable_form = fuzzy_system['fuzzy_forms'][out_variable]['values'][out_variable_value]
        out_variable_form = out_variable_value
        
        # out_variable_form_with_strength = np.tile(np.array([1, 2, 3]), (10, 1))
        out_variable_form_with_strength = (np.tile(out_variable_form, (len(rule_strength), 1)).T * rule_strength).T
        if out_variable in res:
            res[out_variable] = np.maximum(out_variable_form_with_strength, res[out_variable])
        else:
            res[out_variable] = out_variable_form_with_strength

    return res
    # return aggregation(res)


def plot_fuzzy_forms(fuzzy_system):
    for fuzzy_form_name, fuzzy_form in fuzzy_system['fuzzy_forms'].items():
        plt.title(fuzzy_form_name)

        for value_name, value in fuzzy_form['values'].items():
            x = np.linspace(fuzzy_form['domain'][0], fuzzy_form['domain'][1], len(value))
            plt.plot(x, value, label=value_name)

        plt.legend()
        plt.xlim(fuzzy_form['domain'][0], fuzzy_form['domain'][1])
        plt.show()


def run_fuzzy_system(fuzzy_system, variables):
    fuzzy_variables = fuzzify(fuzzy_system, variables)
    res = run_fuzzy_rules(fuzzy_system, fuzzy_variables)
    return centroid_aggregation(fuzzy_system, res)




def simple_rule(variables):
    return and_rule(variables['temp_low'], variables['humidity_normal']), 'speed',  triangle_form([1, 100], 100, 1, 1, 60),#'low'

def simple_rule2(variables):
    return and_rule(variables['temp_high'],  variables['humidity_low']), 'speed', triangle_form([1, 100], 100, 40, 90, 100), # 'high'

fuzzy_system = {
    'rules' : [simple_rule, simple_rule2],
    'aggregation': centroid_aggregation,
    'fuzzy_forms': {
       'temp': {
          'domain': [0, 2],
          'values': {
            'low': triangle_form([0, 2], 100, 0, 0, 0.3),
            'medium': triangle_form([0, 2], 100, 0.2, 0.6, 0.8),
            'high': triangle_form([0, 2], 100, 0.65, 1.9, 2)
          }
        },

        'humidity': {
            'domain': [0, 100],
            'values': {
                'low': triangle_form([0, 100], 100, 0, 10, 70),
                'normal': triangle_form([0, 100], 100, 49, 80, 100)
            }
        },
        # 'speed': {
        #     'domain': [1, 100],
        #     'values': {
        #         'low': triangle_form([1, 100], 100, 1, 1, 60),
        #         'high': triangle_form([1, 100], 100, 40, 90, 100),
        #     }
        # }
    },
}

run_fuzzy_system(fuzzy_system, {
    'temp': np.array([0.1, 0.8]),
    'humidity': np.array([70, 10])
}
)
# print(fuzzy_variables)

# plot_fuzzy_forms(fuzzy_system)

# plt.plot(res['speed'][0])
# plt.plot(res['speed'][1])

# plt.plot(res['speed'][0])
# plt.ylim(0, 1)
# plt.plot(res['speed'][1])
# plt.plot(res['speed'])
```

```python

def task_to_fuzzy_system(task):
    fuzzy_system_words = task['fuzzy_system_words']
    density = task['density']
    def convert_to_triangle(word):
        return word.replace('triangle(', 'triangle_form((0., 1.),' + str(density) +",")

    fuzzy_system = {
        'rules': [],
        'aggregation': eval(fuzzy_system_words['aggregation'] + '_aggregation'),
        'fuzzy_forms': {}
    }

    for (rule_word, rule_out_variable, rule_out_variable_value) in fuzzy_system_words['rules']:
        rule = rule_word.replace('and(', 'and_rule(').replace('or(', 'or_rule(')
        for fuzzy_form_name, fuzzy_form_values in fuzzy_system_words['fuzzy_forms'].items():
            for form_value_name in fuzzy_form_values:
                rule = rule.replace('`' + fuzzy_form_name + '_' + form_value_name + '`', 
                    'variables["' + fuzzy_form_name + '_' + form_value_name + '"]')

        rule_out_form = convert_to_triangle(rule_out_variable_value)
        rule_fn = eval("lambda variables : (" + rule +", '" + rule_out_variable + "', " +  rule_out_form + ")")
        fuzzy_system['rules'].append(rule_fn)

    # normalize_domain = {}
    for fuzzy_form_name, fuzzy_form_valus in fuzzy_system_words['fuzzy_forms'].items():
        # if fuzzy_form_name != 'y':
        #     normalize_domain[fuzzy_form_name] = (
        #         np.min(variables[fuzzy_form_name]),
        #         np.max(variables[fuzzy_form_name])
        #     )
        # else:
        #     normalize_domain[fuzzy_form_name] = (0., 1.)

        fuzzy_system['fuzzy_forms'][fuzzy_form_name] = {
            'domain': (0., 1.),
            'values': {
            }
        }
        
        for fuzzy_form_value_name, fuzzy_form_word in fuzzy_form_valus.items():
            word = convert_to_triangle(fuzzy_form_word)
            fuzzy_system['fuzzy_forms'][fuzzy_form_name]['values'][fuzzy_form_value_name] = \
                eval(word)

    return fuzzy_system

def run_fuzzy_task(task, variables):
    fuzzy_system = task_to_fuzzy_system(task)

    # plot_fuzzy_forms(fuzzy_system)
    return run_fuzzy_system(fuzzy_system, variables)

from sklearn.metrics import mean_squared_error
def run_fuzzy_task_with_loss(fuzzy_task, variables, true_y):
    res = run_fuzzy_task(fuzzy_task, variables)
    pred_y = res['y']
    # print(mean_squared_error(true_y, pred_y))
    return mean_squared_error(true_y, pred_y)
    # return np.sum(np.abs(pred_y - true_y))

run_fuzzy_task_with_loss({
    'density': 15,
    'fuzzy_system_words' : {
        'rules': [
            ('and(`radius_mean_0.0`, or(`texture_mean_0.5`, `texture_mean_1.0`))', 'y', 'triangle(0, 0, 0.6)')
        ],
        'aggregation': 'centroid',
        'fuzzy_forms': {
            'radius_mean': {
                '0.0': 'triangle(0, 0, 0.3)',
                '0.5': 'triangle(0.2, 0.6, 0.8)',
                '1.0': 'triangle(0.7, 0.9, 1.0)'
            },
            'texture_mean': {
                '0.0': 'triangle(0, 0, 0.3)',
                '0.5': 'triangle(0.2, 0.6, 0.7)',
                '1.0': 'triangle(0.65, 0.8, 1.0)'
            },
        }
    }, 
}, variables, y)
```

```python
import pygad
import tqdm

def train_fuzzy_system(variables, columns, values, target_y, num_generations, plot=False):
    gene_space = {"low": 0., "high": 1.}

    num_genes = (2 * 2 + 3) * 4 + (len(columns)) * len(values) * 3

    def ga_solution_to_fuzzy_system_task(solution):
        cur_i = 0
        def next():
            nonlocal cur_i
            val = solution[cur_i]
            cur_i += 1
            return val

        def next_get(llist):
            value = next() * len(llist)
            return llist[int(value)]

        def next_col_val():
            return '`' + next_get(columns) + "_" + next_get(values) + '`'
        
        def next_triangle():
            center = next()
            offset_left = next()
            offset_right = next()
            return 'triangle(' + str(center - offset_left) + ',' + str(center) + ',' + str(center + offset_right) + ')'

        return {
            'density': 15,
            'fuzzy_system_words' : {
                'rules': [
                    ('and(' + next_col_val() + ', ' + next_col_val() + ')', 'y', next_triangle()),
                    ('and(' + next_col_val() + ', ' + next_col_val() + ')', 'y', next_triangle()),
                    ('and(' + next_col_val() + ', ' + next_col_val() + ')', 'y', next_triangle()),
                    ('and(' + next_col_val() + ', ' + next_col_val() + ')', 'y', next_triangle()),
                    # ('or(' + next_col_val() + ', ' + next_col_val() + ')', 'y', next_triangle())
                ],
                'aggregation': 'centroid',
                'fuzzy_forms': {
                    column_name: {
                        value: next_triangle()
                        for value in values
                    }
                    for column_name in (columns)
                },
            }
        }

    def fitness_func(solution, solution_idx):
        task = ga_solution_to_fuzzy_system_task(solution)
        score = run_fuzzy_task_with_loss(task, variables, target_y)
        # print(1./score)
        return 1. / score

    num_generations = num_generations

    num_parents_mating = 50

    sol_per_pop = 100

    init_range_low = 0. # min([min(i) for i in gene_space])
    init_range_high = 0.999 # max([max(i) for i in gene_space])

    parent_selection_type = "sss"
    keep_parents = 1

    crossover_type = "single_point"

    mutation_type = "random"
    mutation_percent_genes = 10


    with tqdm.tqdm(total=num_generations) as pbar:
        ga_instance = pygad.GA(num_generations=num_generations,
                            num_parents_mating=num_parents_mating,
                            fitness_func=fitness_func,
                            sol_per_pop=sol_per_pop,
                            num_genes=num_genes,
                            # gene_type=gene_type,
                            gene_space=gene_space,
                            init_range_low=init_range_low,
                            init_range_high=init_range_high,
                            parent_selection_type=parent_selection_type,
                            keep_parents=keep_parents,
                            crossover_type=crossover_type,
                            mutation_type=mutation_type,
                            mutation_percent_genes=mutation_percent_genes,
                            on_generation=lambda x: pbar.update())#,
                            # save_solutions=True,
                            # save_best_solutions=True)
        ga_instance.run()
    
    if plot:
        ga_instance.plot_fitness()

    solution, solution_fitness, solution_idx = ga_instance.best_solution()
    best_task = ga_solution_to_fuzzy_system_task(solution)
    print("Parameters of the best solution : {solution}".format(solution=best_task['fuzzy_system_words']['rules']))
    print("Fitness value of the best solution = {solution_fitness}".format(solution_fitness=solution_fitness))

    return best_task

columns = list(variables.keys())
values = ['0.0', '0.5', '1.0']
yy = run_fuzzy_task(train_fuzzy_system(variables, columns, values, y, num_generations=20, plot=True), variables)['y']

plt.scatter(X[:, 0], X[:, 1], c=y, alpha=yy)
```
